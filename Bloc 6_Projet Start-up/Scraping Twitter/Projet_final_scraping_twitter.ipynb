```python
url_1 = "https://raw.githubusercontent.com/huynam1012/Projet_jedha/main/companies_data_world.csv"
df1 = pd.read_csv(url_1, encoding='ISO-8859-1')
df1 = df1[df1["status"] == "ipo"]
df1.head(2)
```

```python
df2 = df1.merge(df0, left_on='name', right_on='organization_name')
df2.head(2)
```

```python
df2=  df2[["name", "twitter"]]
df2= df2[df2["twitter"].notnull()]
df2 = df2.drop_duplicates()
df2.head(5)
```

```python
df2['twitter_clean'] = df2['twitter'].str.replace('https://twitter.com/', '')
df2['twitter_clean'] = df2['twitter_clean'].str.replace('https://www.twitter.com/', '')
df2['twitter_clean'] = df2['twitter_clean'].str.replace('http://twitter.com/', '')
df2['twitter_clean'] = df2['twitter_clean'].str.replace('http://www.twitter.com/', '')
df2['twitter_clean'] = df2['twitter_clean'].str.replace('/', '')
companies = df2.copy()
companies['twitter_clean'] = companies['twitter_clean'].drop_duplicates()
```

```python
import tweepy
```

```python
consumer_key = 'mettre votre cl'
consumer_secret = 'mettre votre cl'
access_token = 'mettre votre cl'
access_toket_secret = 'mettre votre cl'
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_toket_secret)
api = tweepy.API(auth)
```

```python
#new data frame to stock tweets
print('Creating df_tweets')
new_columns = ['Twitter_Account', 'Tweeted']
datatweets = pd.DataFrame(columns = new_columns)
print('df_tweets created')
```

```python
#new data frame to stock finance companies errors
print('Creating df_errors')
new_column = ['Twitter_Account']
errors = pd.DataFrame(columns = new_column)
print('df_errors created')
```

```python
number_of_tweet = 5

# Scraping

print('Scrap twitter')
for i in companies['twitter_clean']:
    try:
        for j in tweepy.Cursor(api.user_timeline, screen_name=i, tweet_mode='extended').items(number_of_tweet):
            datatweets= datatweets.append({'Twitter_Account':i,'Tweeted':j.full_text}, ignore_index=True)

    except:
        errors = errors.append({'Twitter_Account':i},ignore_index=True)
        
        continue
        
print('Scraping ended !')
```
